````python
def longest_substring(s):
    # Dictionary to store character positions
    char_pos = {}
    max_length = 0
    start = 0
    longest_substr = ''
    
    for i, char in enumerate(s):
        # If character is seen and its last position is >= start
        if char in char_pos and char_pos[char] >= start:
            # Move start pointer to position after the last occurrence
            start = char_pos[char] + 1
        else:
            # Update max_length if current window is longer
            if i - start + 1 > max_length:
                max_length = i - start + 1
                longest_substr = s[start:i+1]
        
        # Update character position
        char_pos[char] = i
    
    return longest_substr, max_length

# Test the function
s = 'abccdabcdccd'
substring, length = longest_substring(s)
print(f"Longest substring: {substring}")
print(f"Length: {length}")
````

````python
from fastapi import FastAPI, Request, Body
from fastapi.responses import HTMLResponse
from fastapi.templating import Jinja2Templates
import requests
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import json

app = FastAPI()
templates = Jinja2Templates(directory="templates")

class StockAnalyzer:
    def __init__(self, symbol):
        self.symbol = symbol
        self.stock = yf.Ticker(symbol)
        
    def fetch_technical_data(self):
        """Fetch technical indicators and market data"""
        # Get historical data for past year
        hist = self.stock.history(period="1y")
        
        # Calculate technical indicators
        data = {
            'Current_Price': round(hist['Close'].iloc[-1], 2),
            '52W_High': round(hist['High'].max(), 2),
            '52W_Low': round(hist['Low'].min(), 2),
            '50_Day_MA': round(hist['Close'].rolling(window=50).mean().iloc[-1], 2),
            '30_Day_MA': round(hist['Close'].rolling(window=30).mean().iloc[-1], 2),
            '200_Day_MA': round(hist['Close'].rolling(window=200).mean().iloc[-1], 2),
            'Volume': round(hist['Volume'].iloc[-1], 2),
            'Avg_Volume_3M': round(hist['Volume'].iloc[-63:].mean(), 2),
            'Price_Change_1M': round((hist['Close'].iloc[-1] / hist['Close'].iloc[-22] - 1) * 100, 2),
        }
        return data, hist
    
    def create_stock_graph(self):
        """Create an interactive stock price graph"""
        _, hist = self.fetch_technical_data()
        
        fig = go.Figure()
        
        # Add candlestick
        fig.add_trace(go.Candlestick(
            x=hist.index,
            open=hist['Open'],
            high=hist['High'],
            low=hist['Low'],
            close=hist['Close'],
            name='OHLC'
        ))
        
        # Add moving averages
        fig.add_trace(go.Scatter(
            x=hist.index,
            y=hist['Close'].rolling(window=50).mean(),
            name='50 MA',
            line=dict(color='orange')
        ))
        
        fig.add_trace(go.Scatter(
            x=hist.index,
            y=hist['Close'].rolling(window=200).mean(),
            name='200 MA',
            line=dict(color='green')
        ))
        
        fig.update_layout(
            title=f'{self.symbol} Stock Price',
            yaxis_title='Stock Price (USD)',
            xaxis_title='Date',
            template='plotly_dark'
        )
        
        print (fig.to_json())

@app.get("/", response_class=HTMLResponse)
async def index(request: Request):
    return templates.TemplateResponse("index.html", {"request": request})

@app.post("/search")
def search_stock(query: str = Body(...)):
    url = "https://query2.finance.yahoo.com/v1/finance/search"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
    }
    params = {
        "q": query
    }

    try:
        response = requests.get(url, params=params, headers=headers)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        return {"error": str(e)}

@app.post("/stock_data/{symbol}")
def get_stock_data(symbol: str):
    try:
        analyzer = StockAnalyzer(symbol)
        technical_data, _ = analyzer.fetch_technical_data()
        graph_data = analyzer.create_stock_graph()
        
        return {
            "technical_data": technical_data,
            "graph_data": json.loads(graph_data)
        }
    except Exception as e:
        return {"error": str(e)}
````

````python
import yfinance as yf
import pandas as pd

class StockAnalyzer:
    def __init__(self, symbol, period=1):
        self.symbol = symbol
        self.stock = yf.Ticker(symbol)
        self.period = period
    
    def get_monthly_candle_data(self):
        """Return a list of dictionaries for each month's open, high, low, and close values"""
        hist = self.stock.history(period=f"{self.period}y", interval="1mo")
        candles = []
        
        for i, row in hist.iterrows():
            month = i.strftime('%b')  # Short month name
            open_price = row['Open']
            high = row['High']
            low = row['Low']
            close = row['Close']
            bullish = close > open_price
            candles.append({
                'month': month,
                'open': open_price,
                'high': high,
                'low': low,
                'close': close,
                'bullish': bullish
            })
        
        return candles

# Usage example
analyzer = StockAnalyzer('AAPL', period=1)
monthly_candle_data = analyzer.get_monthly_candle_data()

graph_code = """
<!DOCTYPE html>
<html>
<head>
  <style>
    /* Styles are unchanged */
  </style>
</head>
<body>
  <div class="chart-container">
    <div class="y-axis"></div>
    <div class="x-axis"></div>
    <!-- Grid lines and labels (for simplicity, keeping these fixed) -->
    <div class="grid-line" style="bottom: 440px"></div>
    <div class="grid-line" style="bottom: 340px"></div>
    <div class="grid-line" style="bottom: 240px"></div>
    <div class="grid-line" style="bottom: 140px"></div>
    <div class="grid-line" style="bottom: 40px"></div>

    <div class="y-labels">
      <span>50</span>
      <span>45</span>
      <span>40</span>
      <span>35</span>
      <span>30</span>
    </div>

    <div class="x-labels">
      {{months}}
    </div>

    <div class="chart">
      {{candles}}
    </div>

    <span class="axis-label y-title">Price ($)</span>
    <span class="axis-label x-title">Months</span>
  </div>
</body>
</html>
"""
def generate_candle_divs(candle_data):
    candles_html = ""
    months_html = ""
    
    for candle in candle_data:
        height_wick = (candle['high'] - candle['low']) * 10  # Scaling factor
        bottom_wick = candle['low'] * 10
        height_body = abs(candle['close'] - candle['open']) * 10
        bottom_body = min(candle['open'], candle['close']) * 10
        body_class = "bullish" if candle['bullish'] else "bearish"
        
        candles_html += f"""
        <div class="candle">
            <div class="wick" style="height: {height_wick}px; bottom: {bottom_wick}px"></div>
            <div class="body {body_class}" style="height: {height_body}px; bottom: {bottom_body}px"></div>
        </div>
        """
        months_html += f"<span>{candle['month']}</span>"
    
    return candles_html, months_html

# Generate HTML
candles, months = generate_candle_divs(monthly_candle_data)
graph_code = graph_code.replace("{{candles}}", candles).replace("{{months}}", months)
print(graph_code)

````

````python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        num_map = {}  # Dictionary to store the index of numbers
        for i, num in enumerate(nums):
            complement = target - num
            print(complement,'>>')
            if complement in num_map:
                return [num_map[complement], i]
            num_map[num] = i
            print(num_map)
        return []  # Return an empty list if no solution is found
````
